```python
import pandas as pd
from sqlalchemy import create_engine

# DB REQUIREMENTS
username = 'postgres'
password = 'postgres'
host = '192.168.20.11' 
port = '5432'
database = 'workshop'

# CONNECTION
connection = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}'

engine = create_engine(connection)

# SQL query to extract the data
query = 'SELECT * FROM raw_population'

# Load data into pandas DataFrame using the engine
df = pd.read_sql_query(query, engine)

# Preview the data
print(df.head())
```

```python
df.head()
```

```python
df = df.rename(columns={'country/territory': 'country'})

df.head()
```

```python
# Check the sum of the columns with NAN values
df.isna().sum()
```

```python
# Drop NAN values (if there are any)
df.dropna() 
```

```python
# Check for whitespaces on all the columns
whitespace_check = df.applymap(lambda x: x.isspace() if isinstance(x, str) else False)

print(whitespace_check)
```

```python
transformed_df = df

transformed_df.head()
```

```python
# LOADS the transformed data into the DB
transformed_df.to_sql('cleaned_population', engine, if_exists='replace', index=False)

print('New DF is loaded to the DB')
```

```python
transformed_df.head()
```

```python
transformed_df = df.drop(columns=['rank', 'cca3', 'capital',
                                               'continent', '2022_population', '2015_population',
                                               'area_km_squared', 'density_per_km_squared', 'growth_rate',
                                               'world_population_percentage', 'id'])

# Display the first few rows of the transformed DataFrame
transformed_df.head()

```

**CONSUME CSV**

```python
df_population_per_year = transformed_df.melt(id_vars='country', 
                  var_name='year', 
                  value_name='population')

# Clean the 'year' column by removing the "_population" suffix
df_population_per_year['year'] = df_population_per_year['year'].str.replace('_population', '')

# Optionally, convert year to integer
df_population_per_year['year'] = df_population_per_year['year'].astype(int)

# Sort the data if needed
df_population_per_year = df_population_per_year.sort_values(by=['country', 'year']).reset_index(drop=True)
```
